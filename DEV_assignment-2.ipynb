```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('winequality-red.csv')
df.head(500)
```

```python
x=df.density
plt.title('Line Plot')
plt.plot(x)
```

```python
plt.title('Bar Plot')
plt.bar(df.alcohol,df.quality,color='yellow')
```

```python
plt.title('Scatter Plot')
plt.scatter(df.density,df.residual_sugar,color='green')
```

```python
plt.title('Histogram')
plt.hist(df.volatile_acidity,color='red')
```

```python
plt.title('Pie Plot')
plt.pie(df.free_sulfur_dioxide)
```

```python
plt.title('Box Plot')
plt.boxplot(df.fixed_acidity)
```

```python
plt.title('Lollipop Plot')
plt.stem(df.residual_sugar)
```

```python
plt.title('Stack Plot')
plt.stackplot(df.quality,df.alcohol,color='black')
```
